 # Data file generator
 A simple script that writes to a blank data file. The inputs are limited to those which are important for emulating lab conditions, such as: $$\nu,\kappa,\beta_S,f,\rho_0,u,u_t,\omega,dt,n,t_{\text{bin}},dY,dX$$ 

Here we load in the blank data file, read in the rows and make changes as necessary.

```python
filename = 'data_default'
with open(filename,'-r') as fp:
    lines = fp.readlines()
    
    variables = ['Viscosity']
        
    fp.write(line)
```
